```python
# import os
from pandas import DataFrame, read_csv
import numpy as np
from math import log
import matplotlib.pyplot as plt
import pandas as pd
import sys
import matplotlib
import random
%matplotlib inline
```

```python
d1 = pd.read_csv("./timesEj1NC.csv")
NvsT = d1
NvsT = d1.groupby('Size')


dyn = NvsT.mean().plot.line(fontsize=13, figsize=(13,8), logy=False, grid=True)
dyn.set_title("Ej1 - Completo(K=15)", fontsize=21)
dyn.set_ylabel("Tiempo (s)", fontsize=16)
dyn.set_xlabel("#Vértices", fontsize=16)

d1 = pd.read_csv("./timesEj1NnC.csv")
NvsT = d1
NvsT = d1.groupby('Size')


dyn = NvsT.mean().plot.line(fontsize=13, figsize=(13,8), logy=False, grid=True)
dyn.set_title("Ej1 - Fuertemente Conexo(K=15)", fontsize=21)
dyn.set_ylabel("Tiempo (s)", fontsize=16)
dyn.set_xlabel("#Vértices", fontsize=16)
```
