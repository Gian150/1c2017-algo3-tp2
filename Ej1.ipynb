```python
# import os
from pandas import DataFrame, read_csv
import numpy as np
from scipy.optimize import curve_fit
from scipy.stats import pearsonr
from math import log
import matplotlib.pyplot as plt
import pandas as pd
import sys
import matplotlib
import random
%matplotlib inline
```

```python
def cuadrt(x, a, b, c):
    return a * (x**2) + c
def lineal(x, a, b, c):
    return a * x + c
```

# #Cantidad de VÃ©rtices

```python
d1 = pd.read_csv("./data/timesEj1ListNC.csv")
d2 = pd.read_csv("./data/timesEj1MatrixNC.csv")
NvsT = d1
NvsT = d1.groupby('Size')


dyn = NvsT.mean().plot.line(fontsize=13, figsize=(13,8), logy=False, grid=True)
dyn.set_title("Ej1 - Completo(K=15)", fontsize=21)
dyn.set_ylabel("Tiempo (s)", fontsize=16)
dyn.set_xlabel("#VÃ©rtices", fontsize=16)

d1 = pd.read_csv("./data/timesEj1ListNnC.csv")
d2 = pd.read_csv("./data/timesEj1MatrixNnC.csv")
NvsT = d1
NvsT = d1.groupby('Size')


dyn = NvsT.mean().plot.line(fontsize=13, figsize=(13,8), logy=False, grid=True)
dyn.set_title("Ej1 con M=2*(N-1) y K=15)", fontsize=21)
dyn.set_ylabel("Tiempo (s)", fontsize=16)
dyn.set_xlabel("#VÃ©rtices", fontsize=16)
```

```python
d1 = pd.read_csv("./data/timesEj1ListNC.csv")
d2 = pd.read_csv("./data/timesEj1MatrixNC.csv")

xdataMatx = d2['Size']
xdataList = d1['Size']
ydataMatx = d2['Â¬sany']
ydataList = d1['Â¬sany']

plt.plot(xdataMatx, ydataMatx, '.', alpha=.2, label='Matriz Ady')
plt.plot(xdataList, ydataList, '.', alpha=.2, label='Listas Ady')

popt, pcov = curve_fit(cuadrt, xdataMatx, ydataMatx)
yfitMatx = cuadrt(xdataMatx, *popt)
plt.plot(xdataMatx, yfitMatx, 'b-', label='Matriz Ady')

popt, pcov = curve_fit(cuadrt, xdataList, ydataList)
yfitList = cuadrt(xdataList, *popt)
plt.plot(xdataList, yfitList, 'r-', label='Listas Ady')


corrMatx, pvalMatx = pearsonr(yfitMatx,ydataMatx)
corrList, pvalList = pearsonr(yfitList,ydataList)

plt.text(67.2, 0.176, r'Matrix (blue)')
plt.text(67.2, 0.163, r'corr = %.6f'%corrMatx)
plt.text(67, 0.15, r'pval = %.2e'%pvalMatx)

plt.text(110.2, 0.03, r'List (orange)')
plt.text(110.2, 0.02, r'corr = %.6f'%corrList)
plt.text(110, 0.01, r'pval = %.2e'%pvalList)

plt.xlabel('#VÃ©rtices')
plt.ylabel('Tiempo (s)')
plt.title('Fit de cuadrÃ¡tica: Matriz Ady vs Listas Ady')
plt.show()
```
