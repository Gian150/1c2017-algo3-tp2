```python
# import os
from pandas import DataFrame, read_csv
import numpy as np
from math import log
import matplotlib.pyplot as plt
import pandas as pd
import sys
import matplotlib
import random
%matplotlib inline
```

```python
d1 = pd.read_csv("./timesEj1N.csv")
NvsT = d1
NvsT = d1.groupby('Size')

dyn = NvsT.mean().plot.line(fontsize=13, figsize=(13,8), logy=False, grid=True)
dyn.set_title("Tiempos de ej1 (grafo completo)", fontsize=21)
dyn.set_ylabel("Time (s)", fontsize=16)
dyn.set_xlabel("Size", fontsize=16)
```
